```python
flux_steps = 1000
phi_range = np.linspace((0)*2*np.pi, (1)*np.pi, flux_steps)

phi_range_0 = []
phi_range_1 = []
phi_range_2 = []

phi_range_02 = []
phi_range_12 = []
phi_range_22 = []

for phie_id, phie_value in enumerate(phi_range):
    if phie_value < 0.33*np.pi*2:
        phi_range_0.append(0)
        phi_range_1.append((phie_value)/2/np.pi)
        phi_range_2.append((-phie_value)/2/np.pi)
        phi_range_02.append(0)
        phi_range_12.append((phie_value)/2)
        phi_range_22.append((-phie_value)/2)
    else:
        phi_range_0.append(np.pi/np.pi)
        phi_range_1.append((phie_value/2+np.pi/2)/np.pi)
        phi_range_2.append((-phie_value/2+np.pi/2)/np.pi)
        phi_range_02.append(np.pi)
        phi_range_12.append((phie_value/2+np.pi/2))
        phi_range_22.append((-phie_value/2+np.pi/2))
    
    #print phie_id
    #print('φx1_id: {0}, φx2_id: {1}, gap: {2}'.format(phix1_id, phix2_id, eigenenergies[1]-eigenenergies[0]))
```
